```python
class Elevator():
    
    def __init__(self):
        self.current_floor = 1
        self.direction = None
        self.floor_count = 6
        self.queue_up = []
        self.queue_down = []
        
    def move(self):
        if self.direction == 'up':
            self.current_floor+=1
        elif self.direction == 'down':
            self.current_floor-=1
        print('elevator moving', self.direction)
        print('current floor:', self.current_floor)
        print('next stop:', self.next_stop)
        
    def run_until_next_stop(self):
        if not self.direction:
            self._determine_direction()
        self._determine_next_stop()
        while (self.current_floor != self.next_stop):
            self.move()
            if self.current_floor == self.next_stop:
                self._stop_elevator()
        
    def _stop_elevator(self):
        self._remove_from_queue()
        self.direction = None
        print('elevator stopped')
            
    def _determine_next_stop(self):
        if not self.queue_up and not self.queue_down:
            self.direction = None
            print('No calls in queue')
        elif self.direction == 'up':
            if len(self.queue_up) > 0 and min(self.queue_up) > self.current_floor:
                self.next_stop = min(self.queue_up)
            else:
                self.next_stop = max(self.queue_down)
                if self.next_stop < self.current_floor:
                    self.direction = 'down'
        elif self.direction == 'down':
            if len(self.queue_down) > 0 and max(self.queue_down) < self.current_floor:
                self.next_stop = max(self.queue_down)
            else:
                self.next_stop = min(self.queue_up)
                if self.next_stop > self.current_floor:
                    self.direction = 'up'
    
    def _determine_direction(self):
        if self.current_floor == 1:
            self.direction = 'up'
        elif self.current_floor == self.floor_count:
            self.direction = 'down'
        else:
            if len(self.queue_up) > 0 and min(self.queue_up) > self.current_floor:
                self.direction = 'up'
            elif len(self.queue_down) > 0 and max(self.queue_down) < self.current_floor:
                self.direction = 'down'
            else:
                self.direction = 'down'
                self.next_stop = 1
            
    def call(self, passenger_floor, destination_floor):
        if passenger_floor < destination_floor:
            self._add_to_queue(passenger_floor, destination_floor, 'up')
        else:
            self._add_to_queue(passenger_floor, destination_floor, 'down')
    
    def _add_to_queue(self, passenger_floor, destination_floor, passenger_direction):
        if passenger_direction == 'up':
            self.queue_up.append(passenger_floor)
            self.queue_up.append(destination_floor)
            self.queue_up = list(set(self.queue_up))
            self.queue_up.sort()
        elif passenger_direction == 'down':  
            self.queue_down.append(passenger_floor)
            self.queue_down.append(destination_floor)
            self.queue_down = list(set(self.queue_down))
            self.queue_down.sort(reverse=True)
            
    def _remove_from_queue(self):
        if self.current_floor in self.queue_up:
            self.queue_up.remove(self.current_floor)
            print(f'removing {self.current_floor} from queue up')
        elif self.current_floor in self.queue_down:
            self.queue_down.remove(self.current_floor)
            print(f'removing {self.current_floor} from queue down')
            
```
